**EAST ASIA**

```python
# Import
import pandas as pd
import matplotlib.pyplot as plt

# Load
df = pd.read_excel("Trade.xlsx", sheet_name="EA")

# Explore
df.head()
```

```python
### Modify Variables ###

# List of columns to multiply by 100
columns_to_multiply = [
    'Imp_Regional', 'Imp_Europe', 'Imp_US', 'Imp_Other',
    'Exp_Regional', 'Exp_Europe', 'Exp_US', 'Exp_Other'
]

# Multiply the specified columns by 100
df[columns_to_multiply] = df[columns_to_multiply] * 100

# Display the first few rows of the updated DataFrame
df.head()
```

```python
### Imports ###

# Create a line plot
plt.figure(figsize=figsize)

#Plot
plt.plot(df['Year'], df['Imp_Regional'], label="East Asia", linestyle="dashed", color="black")
plt.plot(df['Year'], df['Imp_Europe'], label="Europe", linestyle=(5, (10, 3)), color="dimgray")
plt.plot(df['Year'], df['Imp_US'], label="North America", linestyle = "solid", color="silver")

# Adding labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Share of Imports to East Asia (%)', fontsize=12)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

# Show the plot
plt.grid(False)
plt.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=2,          # Increase ncol for more spread-out entries
    frameon=False,   # Remove the rectangle around the legend
    columnspacing=4,  # Increase spacing between columns
    fontsize=12
)
plt.show()
```

```python
### Exports ###

# Create a line plot
plt.figure(figsize=figsize)

#Plot
plt.plot(df['Year'], df['Exp_Regional'], label="East Asia", linestyle="dashed", color="black")
plt.plot(df['Year'], df['Exp_Europe'], label="Europe", linestyle = (5, (10, 3)), color="dimgray")
plt.plot(df['Year'], df['Exp_US'], label="North America", linestyle = "solid" , color="silver")

# Adding labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('Share of Exports from East Asia(%)', fontsize=12)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

# Show the plot
plt.grid(False)
plt.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=2,          # Increase ncol for more spread-out entries
    frameon=False,   # Remove the rectangle around the legend
    columnspacing=4,  # Increase spacing between columns
    fontsize=12
)
plt.show()
```
